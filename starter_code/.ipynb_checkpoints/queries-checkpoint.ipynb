# Answers
### 1. All the companies that it's name match 'Babelgum'. Retrieve only their `name` field.

```python
list(db.companies.find({'name':'Babelgum'},{'name':1, '_id': 0}))
```

### 2. All the companies that have more than 5000 employees. Limit the search to 20 companies and sort them by **number of employees**.

```python
list(db.companies.find(filter= {'number_of_employees': {'$gt':5000}},
sort = [('number_of_employees',1)], limit = 20, projection={'name':1,'_id':0}))
```

### 3. All the companies founded between 2000 and 2005, both years included. Retrieve only the `name` and `founded_year` fileds.

```python
list(db.companies.find(filter={'$and':[{'founded_year':{'$gte':2000}}, {'founded_year':{'$lte':2005}}]},
limit = 20, projection = {'founded_year':1, 'name':1, '_id':0}))
```

### 4. All the companies that had a Valuation Amount of more than 100.000.000 and have been founded before 2010. Retrieve only the `name` and `ipo` fields.

```python
list(db.companies.find({'$and': [{'founded_year': {'$lt': 2010}},
    {"ipo.valuation_amount": {'$gt': 100000000}}]}, projection = {'ipo.valuation_amount': 1, 'name':1, '_id':0}))
```

### 5. All the companies that have less than 1000 employees and have been founded before 2005. Order them by the number of employees and limit the search to 10 companies.

```python
list(db.companies.find(filter={'$and':[{'number_of_employees': {'$lt': 1000}}, {'founded_year': {'$lt': 2005}}]},
     sort = [('number_of_employees', -1)], limit = 10, projection= {'name': 1, 'number_of_employees': 1, '_id': 0}))
```

### 6. All the companies that don't include the `partners` field.

```python
list(db.companies.find(filter = {'partners':{'$exists':False}}, 
                       projection= {'name': 1, '_id':1}))
```

### 7. All the companies that have a null type of value on the `category_code` field.

```python
list(db.companies.find(filter = {'category_code':{'$eq':None}},
                       projection={'name':1,'_id':0}))
```

### 8. All the companies that have at least 100 employees but less than 1000. Retrieve only the `name` and `number of employees` fields.

```python
list(db.companies.find(filter = {'$and':[{'number_of_employees':{'$gte':100}},{'number_of_employees':{'$lt':1000}}]},
    projection = {'name':1, '_id':0, 'number_of_employees':1}))
```

### 9. Order all the companies by their IPO price descendently.

```python
list(db.companies.find(filter = {'ipo.valuation_amount': {'$gte':1}},
    projection = {'name':1, '_id':0, 'ipo.valuation_amount':1},
    sort = [("ipo.valuation_amount",-1)]))
```

### 10. Retrieve the 10 companies with more employees, order by the `number of employees`

```python
list(db.companies.find(filter = {'number_of_employees': {'$exists': True}},
projection = {'name': 1, '_id': 0, 'number_of_employees': 1}, limit = 10,
sort = [('number_of_employees', -1)]))
```

### 11. All the companies founded on the second semester of the year. Limit your search to 1000 companies.

```python
list(db.companies.find(filter = {'founded_month': {'$gt': 6}}, limit = 1000,
projection ={'name': 1, '_id': 0, 'founded_month':1}))
```

### 13. All the companies founded before 2000 that have and acquisition amount of more than 10.000.000

```python
list(db.companies.find(filter =  {'$and': [{'founded_year': {'$lt':2000}},
                                {"acquisition.price_amount": {'$gte':10000000}}]},
projection = {'name': 1, '_id':0, 'founded_year': 1, "acquisition.price_amount":1}))
```

### 14. All the companies that have been acquired after 2015, order by the acquisition amount, and retrieve only their `name` and `acquisiton` field.

```python
 list(db.companies.find(filter = {"acquisition.acquired_year": {'$gt': 2015}},
                       projection = {'name': 1, '_id':0, "acquisition.acquired_year": 1, "acquisition.price_amount":1},
                       sort = [('acquisition.price_amount',1)]))
```

### 15. Order the companies by their `founded year`, retrieving only their `name` and `founded year`.

```python
list(db.companies.find(filter = {'founded_year': {'$exists': True}}, limit = 100,
                     projection = {'name': 1, '_id':0, 'founded_year':1}, 
                     sort = [('founded_year', -1)]))
# Se usa limit 100 porque fueron demasiado registros
```

### 16. All the companies that have been founded on the first seven days of the month, including the seventh. Sort them by their `aquisition price` descendently. Limit the search to 10 documents.

```python
list(db.companies.find(filter = {'founded_day': {'$lte': 7}},
                       projection = {'name': 1, '_id': 0, 'founded_day': 1, "acquisition.price_amount": 1},
                      limit = 10, sort = [("acquisition.price_amount", -1)]))
```

### 17. All the companies on the 'web' `category` that have more than 4000 employees. Sort them by the amount of employees in ascendant order.

```python
list(db.companies.find(filter = {'$and':[{'category_code': 'web'}, {'number_of_employees':{'$gt': 4000}}]},
                      projection = {'name': 1, '_id': 0, 'number_of_employees': 1},
                      sort = [('number_of_employees',1)]))
```

### 18. All the companies which their acquisition amount is more than 10.000.000, and currency are 'EUR'.


```python
list(db.companies.find(filter = {'$and':[{'acquisition.price_currency_code': 'EUR'}, {'acquisition.price_amount':{'$gt': 10000000}}]},
                      projection = {'name': 1, '_id': 0, 'acquisition.price_amount': 1}))
```

### 19. All the companies that have been acquired on the first trimester of the year. Limit the search to 10 companies, and retrieve only their `name` and `acquisition` fields.

```python
list(db.companies.find(filter = {"acquisition.acquired_month": {'$lte':3}},
                       limit = 10, projection = {'name':1, '_id': 0, "acquisition.acquired_month":1}))
```

### 20. All the companies that have been founded between 2000 and 2010, but have not been acquired before 2011.

```python
list(db.companies.find(filter = {'$and':[{'founded_year': {'$gte': 2000}}, {'founded_year':{'$lte': 2010}}, {"acquisition.acquired_year":{'$gt': 2011}}]},
                       projection = {'name':1, '_id': 0, 'founded_year':1, "acquisition.acquired_year":1}))
```
